# Pre-processing the dataset


```python
df.isnull().sum()  # checking if there's any null values
```

```python
# checking unique values

df.apply(lambda x: len(x.unique()))
```

```python
# Changing attributes name

df.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'Latitude',
    'longitude':'Longitude','summary':'Summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill','nwound':'Wound'},inplace=True)
    
```

```python
df = df[['Year','Month','day','Country','State','Region','City','Latitude','Longitude',"Attacktype",'kill',
               'Wound','target1','Summary','Group','Targettype','Weapon','motive']]
```

# Analysing the dataset

```python
plt.figure(figsize= (20,5))

df[['Group','kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('kill',ascending=False).head(10).plot(kind='bar',color='cyan')


plt.title("Top 10 Terrorist Groups")
plt.xlabel("Terrorist Group")
plt.ylabel("People killed")
plt.grid(True)
plt.show()
```

```python
x = df["Year"].unique()
y = df["Year"].value_counts().sort_index()
plt.figure(figsize = (20,10))
sns.barplot(x ,y,palette = "rocket")        
plt.title("Attacks in years")
plt.xlabel("Year of Attack")
plt.ylabel("Number of Attacks")
plt.xticks(rotation =90)
plt.show()
```

```python
plt.subplots(figsize=(10,5))
sns.barplot(df["Country"].value_counts()[:10].index,df["Country"].value_counts()[:10].values,palette="pastel")
plt.title("Top Countries Affected")
plt.xlabel("Countries")
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.subplots(figsize=(10,5))
sns.barplot(df["Targettype"].value_counts()[:10].index,df["Targettype"].value_counts()[:10].values,palette="pastel")
plt.title("Attack Location")
plt.xlabel("Locations")
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

```python
df[['Attacktype','kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',color=['blue'])
plt.xticks(rotation=90)
plt.title("Total Killings")
plt.ylabel('Number of people')
plt.xlabel('Attack type')
plt.show()
```

```python
corr = df.corr()  # corr = variable $ df.corr is the correlation function for it

plt.figure(figsize= (5,10))
sns.heatmap(corr,annot= True, cmap = 'coolwarm' )
```
